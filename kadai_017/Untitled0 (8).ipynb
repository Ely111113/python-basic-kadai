```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def check_adult(self):

    if self.age >= 20:
      print("成人")
    else:
      print("未成年")

human1 = Human("太郎",15)
human2 = Human("治郎",25)
human3 = Human("侍太郎",59)

humans = []

humans.append(human1)
humans.append(human2)
humans.append(human3)

i = 0

for i in range(len(humans)):
  humans[i].check_adult()
```
